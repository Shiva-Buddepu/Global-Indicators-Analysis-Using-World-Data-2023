#Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('world-data-2023.csv')
```

```python
data
```

#Comparative Analysis of GDP, Life Expectancy, and Unemployment Rate

```python
# Select relevant columns for comparative analysis
columns = ['Country', 'GDP', 'Life expectancy', 'Unemployment rate']
comparative_data = data[columns]
comparative_data.to_csv('comparative_data.csv', index=False)
```

```python
comparative_data
```

```python
# Calculate descriptive statistics for each column
statistics = comparative_data.describe()
statistics.to_csv('statistics.csv', index=False)
```

```python
statistics
```

```python
# Compare countries based on GDP
gdp_ranking = comparative_data.sort_values('GDP', ascending=False)
gdp_ranking.to_csv('gdp_ranking.csv', index=False)
```

```python
gdp_ranking
```

```python
# Compare countries based on life expectancy
life_expectancy_ranking = comparative_data.sort_values('Life expectancy', ascending=False)
life_expectancy_ranking.to_csv('life_expectancy_ranking.csv', index=False)
```

```python
life_expectancy_ranking
```

```python
top30_life_expectancy = life_expectancy_ranking.head(30)
plt.figure(figsize=(15, 6))  # Set the figure size
top30_life_expectancy.plot(x='Country', y='Life expectancy', kind='bar')
plt.title('Top 30 Countries - Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.tight_layout()  # Adjust spacing
plt.savefig('top30_life_expectancy_graph.png')
```

```python
# Compare countries based on unemployment rate
unemployment_ranking = comparative_data.sort_values('Unemployment rate')
unemployment_ranking.to_csv('unemployment_ranking.csv', index=False)
```

```python
unemployment_ranking
```

#Demographic Analysis of Top 20 Countries

```python
# Select relevant columns for demographic analysis
columns = ['Country', 'Birth Rate', 'Infant mortality', 'Urban_population']
demographic_data = data[columns]
```

```python
# Make a copy of the DataFrame to avoid the warning
demographic_data = demographic_data.copy()
```

```python
# Sort data by Birth Rate in descending order
demographic_data.sort_values('Birth Rate', ascending=False, inplace=True)
```

```python
# Take top 20 entries
demographic_data_top20 = demographic_data.head(20)
```

```python
# Analyze birth rate
birth_rate_statistics = demographic_data_top20['Birth Rate'].describe()
birth_rate_statistics.to_csv('birth_rate_statistics.csv')
```

```python
birth_rate_statistics
```

```python
# Generate birth rate graph
demographic_data_top20.plot(x='Country', y='Birth Rate', kind='bar')
plt.title('Top 20 Countries - Birth Rate')
plt.xlabel('Country')
plt.ylabel('Birth Rate')
plt.xticks(rotation=90)
plt.savefig('birth_rate_graph.png')
```

```python
# Analyze infant mortality rate
infant_mortality_statistics = demographic_data_top20['Infant mortality'].describe()
infant_mortality_statistics.to_csv('infant_mortality_statistics.csv')
```

```python
infant_mortality_statistics
```

```python
# Generate infant mortality rate graph
demographic_data_top20.plot(x='Country', y='Infant mortality', kind='bar')
plt.title('Top 20 Countries - Infant Mortality Rate')
plt.xlabel('Country')
plt.ylabel('Infant Mortality Rate')
plt.xticks(rotation=90)
plt.savefig('infant_mortality_graph.png')
```

```python
# Analyze urban/rural distribution
urban_rural_distribution = demographic_data_top20.groupby('Urban_population').size()
urban_rural_distribution.to_csv('urban_rural_distribution.csv')
```

```python
urban_rural_distribution
```

```python
# Generate urban/rural distribution graph
urban_rural_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 20 Countries - Urban vs Rural Population Distribution')
plt.ylabel('')
plt.savefig('urban_rural_distribution_graph.png')
```
